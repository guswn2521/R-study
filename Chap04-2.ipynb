## 다이아몬드 데이터 분석 - 가격에 영향을 미치는 것.

```r
library('ggplot2')
```

```r
df <- diamonds
str(df)
## tibble : 데이터 형태중 하나
```

```r
head(df,3)
```

```r
## [행, 열]
## data 임의의 개수만큼 추출
## sample(a,b):a 개 중 b 개 추출 
Sample <- df[sample(nrow(df), 300),]
attach(Sample)
```

```r
str(Sample)
```

```r
boxplot(price)
## I 위는 이상치 데이터
```

```r
## 가격 이상치 - EDA로 원인찾기
## price~. : price를 y축으로, x축은 모두다. 

par(mfrow = c(3,3))
plot(data=Sample, price~.)
```

```r
## price에 대해 carat, x, y, z등이 영향이 있어 보임.
par(mfrow = c(2,2))
plot(price~carat+x+y+z)
```

```r
## 각 변수간의 상관계수 구하기
Sample2 <- data.frame(price, carat, x, y, z)
cor(Sample2)
## price에 상관관계 : carat > y > x > z 
```

```r
## price, carat, x 특징 파악
par(mfrow = c(1,3))
boxplot(price, main='price')
boxplot(carat, main='carat')
boxplot(x, main='x')
```

```r
## carat 오름차순 정렬
## order 함수
reCarat <- carat[order(carat)]
par(mfrow=c(1,1))
plot(reCarat)
```

carat과 price는 지수적관계  => 이로 인해 이상치가 발생했을거라 예상

```r
## carat 값 급격히 커지는 부분 제거
## subset 함수 : 잘라오기
## {} : 조건
Sample2 <- subset(Sample, subset = {Sample$carat<0.9})
par(mfrow = c(1,2))
boxplot(Sample$carat, main='carat 1.5 포함')
boxplot(Sample2$carat, main = 'carat 1.5 미포함')
```

## 섬세한 그래프 그리기

```r
library('ggthemes')
str(diamonds)
```

```r
## x와 price 산점도 그리기
ggplot(diamonds, aes(x=x, y=price)) + geom_point()
```

```r
## clarity는 price에 영향 없어보임.
ggplot(diamonds, aes(x=x, y=price, colour=clarity)) + geom_point()
```

```r
ggplot(diamonds, aes(x=x, y=price, colour=clarity))+geom_point()+theme_solarized_2()
```

```r
## 데이터 겹침 조절
## alpha - 점을 연하게 찍음
ggplot(diamonds, aes(x=x, y=price, colour=clarity))+geom_point(alpha=0.03) + 
guides(colour=guide_legend(override.aes = list(alpha = 1))) + 
theme_solarized_2()
```

```r
## xlim() : x축 범위 조절
## geom_hline() : 수평선 표시
ggplot(diamonds, aes(x=x, y=price, colour=clarity)) + 
geom_point(alpha=0.03) + 
geom_hline(yintercept = mean(diamonds$price), color = 'turquoise3', alpha = .8) +
xlim(3,9) + 
guides(colour=guide_legend(override.aes = list(alpha = 1))) + 
theme_solarized_2()
```

### dplyr 패키지 - mutate() 

```r
## mutate ~~ cbind()
df <- mutate(df, SexF=ifelse(SexRatio < 1, "여자비율높음", ifelse(SexRatio>1, "남자비율높음", "남여비율같음")))
head(df)
```

```r
df2 <- filter(df, Provinces=='경기도')
graph <- ggplot(df2, aes(x=City, y=(SexRatio-1), fill=SexF)) + 
geom_bar(stat='identity') + 
theme_wsj()
graph
```

### reshape2 패키지 - melt() 
- 명목형 변수로 만들어줌.

```r
install.packages('reshape2')
```

```r
library('reshape2')
```

```r
## 도별 합계
group <- group_by(df, Provinces)
df2 <- summarize(group, SumPopulation=sum(Population), Male=sum(Male), Female=sum(Female))
df2
```

```r
## melt() 명목형 변수로 처리
df3 <- melt(df2, measure.vars=c('Male','Female'))
df2
df3
```

```r
## 컬럼명 바꾸기
colnames(df3)[3] <- 'Sex'
colnames(df3)[4] <- 'Population'
head(df3)
```

```r
## 남여 비율 추가
df4 <- mutate(df3, Ratio=Population/SumPopulation)
df4$Ratio <- round(df4$Ratio, 3)
head(df4)
```

```r
## coord_cartesian():확대
g1 <- ggplot(df4, aes(x=Provinces, y=Ratio, fill=Sex)) + geom_bar(stat='identity') + 
coord_cartesian(ylim=c(0.45, 0.55)) + 
theme_wsj()
g2 <- geom_text(aes(y=Ratio, label=Ratio), colour='white')
g1 + g2
```

```r
df4 <- mutate(df4, Position=ifelse(Sex=='Male',0.475, 0.525))
head(df4)
```

```r
g1 <- ggplot(df4, aes(x=Provinces, y=Ratio, fill=Sex)) + 
geom_bar(stat='identity')+
coord_cartesian(ylim=c(0.45, 0.55)) + theme_wsj()
g2 <- geom_text(aes(y=Position, label=Ratio), colour='white')
g1+g2
```

### 시간에 따른 연령별 인구변화 그래프 

```r
library('scales')
```

```r
df <- read.csv('r-ggagi-data/example_population2.csv')
df <- tbl_df(df)
```

```r
head(df)
```

```r
## 10살 단위 새로운 dataframe생성
group <- group_by(df, Time)
df2 <- summarise(group, s0=sum(age0to4, age5to9),
                s10=sum(age10to14, age15to19),
                s20=sum(age20to24, age25to29),
                s30=sum(age30to34, age35to39),
                s40=sum(age40to44, age45to49),
                s50=sum(age50to54, age55to59),
                s60=sum(age60to64, age65to69),
                s70=sum(age70to74, age75to79),
                s80=sum(age80to84, age85to89),
                s90=sum(age90to94, age95to99),
                s100=sum(age100to104, age105to109))
```

```r
head(df2)
```

```r
df3 <- melt(df2, measure.vars=c('s0','s10','s20','s30','s40','s50','s60','s70','s80','s90','s100'))
tail(df3)
colnames(df3) <- c('Time', 'Generation', 'Population')
```

```r
head(df3)
```

```r
g1 <- ggplot(df3, aes(x=Time, y=Population, colour=Generation, fill=Generation)) +
geom_area(alpha=.6) + theme_wsj()
g1
```

```r
g2 <- ggplot(df3, aes(x=Time, y=Population, colour=Generation, fill=Generation)) +
geom_area(alpha=.4) + theme_wsj()
g2 + scale_y_continuous(labels=comma)
```
